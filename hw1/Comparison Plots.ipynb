```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('paper')
sns.set_style('darkgrid')
```

```python
objs_mean = np.array([-0.1676, -0.1775, -0.1778])
objs_stds = np.array([0.0883, 0.1023, 0.0943])

errs_mean = np.array([0.0163, 0.0191, 0.0165])
errs_stds = np.array([0.0167, 0.0243, 0.017])

times_mean = np.array([2.455, 1.4, 1.095])
times_stds = np.array([0.2748, 0.52, 0.4882])
```

```python
labels = ['IP', 'SQP', 'AS']
```

```python
xs = [0, 1, 2]
```

```python
plt.figure(figsize=(8, 6), dpi=300)
plt.bar(xs, objs_mean, yerr=objs_stds/2, tick_label=labels)

plt.title('Optimization Costs')
plt.ylabel('Costs')

plt.savefig('costs.png')
```

```python
plt.figure(figsize=(8, 6), dpi=300)
plt.bar(xs, errs_mean, yerr=errs_stds/2, tick_label=labels)

plt.title('Pose Errs')
plt.ylabel('Pose Errs')

plt.savefig('errs.png')
```

```python
plt.figure(figsize=(8, 6), dpi=300)
plt.bar(xs, times_mean, yerr=times_stds/2, tick_label=labels)

plt.title('Runtimes')
plt.ylabel('Time (s)')

plt.savefig('times.png')
```
